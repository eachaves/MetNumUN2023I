```python
class stock:
    def __init__(self, ticker, open, close, volume):
        self.ticker = ticker
        self.open = open
        self.close = close
        self.volume = volume
        self.rate_return = float(close)/open - 1
 
    def update(self, open, close):
        self.open = open
        self.close = close
        self.rate_return = float(self.close)/self.open - 1
 
    def print_return(self):
        print(self.rate_return)
```

```python
frutas = stock('mora', 800, 2700, 20109375)
ropa = stock('zapatos', 50000, 65000, 1561616)
```

```python
frutas.ticker
ropa.print_return()
ropa.update(95000,102000)
ropa.print_return()
```
